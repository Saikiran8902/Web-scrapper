```python
import requests
from bs4 import BeautifulSoup
l1=[]
for i in l:
    if str(i).startswith(('http://', 'https://')):
        try:
            response = requests.get(str(i),verify=False)
        except requests.exceptions.ConnectionError as e:
            l1.append(' ')
        else:
            if(str(i)=='https://ammcollege.in/'):
                l1.append(' ')
                continue
            soup = BeautifulSoup(response.text, 'html.parser')
            address = soup.find('div', {'id': 'Address'})
            if(address):
                l1.append(address.text.strip())
            else:
                l1.append(' ')
l1
```

```python
url = 'http://www.useams.org/'

# Send an HTTP request to the URL of the webpage you want to access
response = requests.get(url)

# Parse the content of the webpage with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the element containing the address
address_element = soup.find('div', {'class': 'adr'})

# Extract the text from the element
if address_element is not None:
    address = address_element.get_text(strip=True)
else:
    address = ''

# Print the address
print(address)
```
